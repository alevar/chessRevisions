```python
import pandas as pd
import numpy as np
from itertools import groupby
from operator import itemgetter
import subprocess
import sys
import os
import csv
import matplotlib.pyplot as plt
from matplotlib import animation
import seaborn as sns
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams['font.family'] = "serif"
%matplotlib inline
```

```python
mergedExonDF=pd.read_csv("./revision/mergedExonDF.csv")
mergedExonDF.drop(["Unnamed: 0","seqid","strand","uid"],axis=1,inplace=True)
mergedExonDF
```

```python
resDF=pd.read_csv("./revision/resDF.csv")
resDF.drop("Unnamed: 0",axis=1,inplace=True)
resDF["gID"]=resDF.parent.str.extract('(CHS.(\d)*)',expand=True)[0]
resDF=resDF[["parent","gID","match"]]
resDF
```

```python
# now it is time to split them and group them by
mergedExonDF['tmp0']=mergedExonDF.match.mask(mergedExonDF.match.ne(0))
mergedExonDF['tmp1']=mergedExonDF.match.mask(mergedExonDF.match.ne(1))
mergedExonDF['tmp2']=mergedExonDF.match.mask(mergedExonDF.match.ne(2))

groupedDF=pd.DataFrame(mergedExonDF.groupby("gID").agg({'tmp0':'count','tmp1':'count','tmp2':'count','match':'count'})).reset_index()
print("number of weirdos: ",len(groupedDF[~(groupedDF['match']==(groupedDF['tmp0']+groupedDF['tmp1']+groupedDF["tmp2"]))]))
groupedDF.head()
```

```python
resDF["match"]=np.where(resDF['match']==99999,0,resDF['match'])

# now it is time to split them and group them by
resDF['tmp0']=resDF.match.mask(resDF.match.ne(0))
resDF['tmp1']=resDF.match.mask(resDF.match.ne(1))
resDF['tmp2']=resDF.match.mask(resDF.match.ne(2))

groupedResDF=pd.DataFrame(resDF.groupby("gID").agg({'tmp0':'count','tmp1':'count','tmp2':'count','match':'count'})).reset_index()
print("number of weirdos: ",len(groupedResDF[~(groupedResDF['match']==(groupedResDF['tmp0']+groupedResDF['tmp1']+groupedResDF["tmp2"]))]))
groupedResDF.head()
```

```python
print(groupedResDF.tmp2.sum())
print(groupedResDF.tmp1.sum()+groupedResDF.tmp0.sum())
print(groupedDF.tmp2.sum())
print(groupedDF.tmp1.sum()+groupedDF.tmp0.sum())
```
