```python
import pandas as pd
import numpy as np
import os
import sys

version="2.01"

gff3Cols=["seqid","source","type","start","end","score","strand","phase","attributes"]
```

```python
df=pd.read_csv("./chess2.0.gff",sep="\t",skiprows=3,names=gff3Cols)
df=df[df["seqid"]=="chrM"].reset_index(drop=True)
df=df[df["type"].isin(["gene","exon","transcript"])].reset_index(drop=True)
df
```

```python
gencode=pd.read_csv("/home/sparrow/genomicData/hg38/annotations/hg38_p8_gencode25.gff3",skiprows=5,names=gff3Cols,sep="\t")
gencode=gencode[gencode["seqid"]=="chrM"].reset_index(drop=True)
gencode=gencode[gencode["type"].isin(["gene","exon","transcript","mRNA"])].reset_index(drop=True)
print("number of mitochondrial genes in GENCODE is: "+str(len(gencode[gencode["type"]=="gene"])))
gencode
```

```python
refseq=pd.read_csv("/home/sparrow/genomicData/hg38/annotations/hg38_p8_refseq.gff3",skiprows=5,names=gff3Cols,sep="\t")
refseq=refseq[refseq["seqid"]=="chrM"].reset_index(drop=True)
refseq=refseq[refseq["type"].isin(["exon","tRNA","rRNA"])].reset_index(drop=True)
refseq
```

```python
# now need to count the number of exons per given transcript
# as well as the number of transcripts per given gene
# this should ultimately be 1-1 ratio in both cases

# to get the number of transcripts per gene
chessGT=df[df["type"].isin(["transcript"])].reset_index(drop=True)
chessTE=df[df["type"].isin(["exon"])].reset_index(drop=True)

chessGT["gID"]=chessGT["attributes"].str.split("ID=",expand=True)[1].str.split(";",expand=True)[0]
chessGT["gID"]=np.where(chessGT["type"]=="transcript","CHS."+chessGT["gID"].str.split(".",expand=True)[1],chessGT["gID"])
chessGT=chessGT[["gID","type"]].groupby("gID").count().reset_index()
chessGT.columns=["gID","count"]
display(chessGT)

chessTE["tID"]=chessTE["attributes"].str.split("Parent=",expand=True)[1].str.strip("\n")
chessTE=chessTE[["tID","type"]].groupby("tID").count().reset_index()
chessTE.columns=["tID","count"]
display(chessTE)
```

```python
# same now for gencode
gencodeGT=gencode[gencode["type"].isin(["transcript","mRNA"])].reset_index(drop=True)
gencodeTE=gencode[gencode["type"].isin(["exon"])].reset_index(drop=True)

gencodeGT["gID"]=gencodeGT["attributes"].str.split("geneID=",expand=True)[1].str.split(";",expand=True)[0]
gencodeGT=gencodeGT[["gID","type"]].groupby("gID").count().reset_index()
gencodeGT.columns=["gID","count"]
display(gencodeGT)

gencodeTE["tID"]=gencodeTE["attributes"].str.split("Parent=",expand=True)[1].str.strip("\n")
gencodeTE=gencodeTE[["tID","type"]].groupby("tID").count().reset_index()
gencodeTE.columns=["tID","count"]
display(gencodeTE)
```

```python
# same now for gencode

refseqGT=refseq[refseq["type"].isin(["transcript","tRNA","rRNA"])].reset_index(drop=True)
refseqTE=refseq[refseq["type"].isin(["exon"])].reset_index(drop=True)

refseqGT["gID"]=refseqGT["attributes"].str.split("geneID=",expand=True)[1].str.split(";",expand=True)[0]
refseqGT=refseqGT[["gID","type"]].groupby("gID").count().reset_index()
refseqGT.columns=["gID","count"]
display(refseqGT)

refseqTE["tID"]=refseqTE["attributes"].str.split("Parent=",expand=True)[1].str.strip("\n")
refseqTE=refseqTE[["tID","type"]].groupby("tID").count().reset_index()
refseqTE.columns=["tID","count"]
display(refseqTE)
```

```python
# Seems like both refseq and gencode 25 have the same number of genes and transcripts
# let us now compare the two to see if the coordinates also match between the two
ge=gencode[gencode["type"].isin(["exon"])].reset_index(drop=True)
re=refseq[refseq["type"].isin(["exon"])].reset_index(drop=True)
merged=ge.merge(right=re,how="outer",on=["start","end"],indicator=True)
# display(merged)
print("there were %d out of %d exons with matching coordinates between RefSeq and GENCODE found"%(len(merged[merged["_merge"]=="both"]),len(merged)))
# identical numbers imply that RefSeq and gencode have identical exon annotations of the mitochondrial genome

# now to investigate the missing transcripts from both gencode and refseq
# in gencode the missing transcripts are due to the "mRNA" type present instead of the "transcript" type
# The question is whether refseq and gencode have the same transcripts missing

gt=gencode[gencode["type"]=="mRNA"].reset_index(drop=True) # collection of GENCODE mRNA transcripts
ge["tID"]=ge["attributes"].str.split("Parent=",expand=True)[1].str.strip("\n")
setMRNA=set(gt["attributes"].str.split("ID=",expand=True)[1].str.split(";",expand=True)[0])
gte=ge[ge["tID"].isin(setMRNA)].reset_index(drop=True)

# now to do the sme for refseq
re["tID"]=re["attributes"].str.split("Parent=",expand=True)[1].str.strip("\n")
rt=refseq[~(refseq["type"]=="exon")].reset_index(drop=True)
setRT=set(rt["attributes"].str.split("ID=",expand=True)[1].str.split(";",expand=True)[0])
re=re[~(re["tID"].isin(setRT))].reset_index(drop=True)

# now to figure out whether the two are the same we shall merge the two coordinate frames
regte=re.merge(right=gte,how="outer",on=["start","end"],indicator=True)
print("%d out of %d exons with an ambiguous transcript from RefSeq match ambiguous exons from GENCODE"%(len(regte[regte["_merge"]=="both"]),len(regte)))
```

```python
# now just need to ask Ela/Steven about it
# for now let's get back to the CHESS annotation and see what we can do to improve it further
# 1. why not see how many single-exon transcripts there are in total first
# 2. second we shall see how many of these transcripts/exons match those defined in the gencode and refseq annotations


# first we shall confirm there are no stray exons (such as found in gencode and refseq)
ce=df[df["type"]=="exon"].reset_index(drop=True)
ce["tID"]=ce["attributes"].str.split("Parent=",expand=True)[1].str.strip("\n")
ct=df[df["type"]=="transcript"].reset_index(drop=True)
ct["gID"]=ct["attributes"].str.split("Parent=",expand=True)[1].str.split(";",expand=True)[0]

ceg=ce[["tID","type"]].groupby("tID").count()
ceg.reset_index(inplace=True)
ctg=ct[["gID","type"]].groupby("gID").count()
ctg.reset_index(inplace=True)

good_tIDs=ceg[ceg["type"]==1]["tID"].tolist() # get single-exon only
bad_tIDs=ceg[ceg["type"]>1]["tID"].tolist()
good_gIDs=["CHS."+x.split(".")[1] for x in good_tIDs]
bad_gIDs=["CHS."+x.split(".")[1] for x in bad_tIDs]

# next we need ot see whether these geneIDs account for all genes currently present in chess
# basically, the question is whether all genes have at one single-exon transcript annotated for them
all_gIDs=ctg["gID"].tolist()
all_gIDs
print("%d genes have a transcript with a single exon out of a total of %d mitochondrial genes present in chess"%(len(good_gIDs),len(all_gIDs)))
```

```python
# Now we need to find out which ones from gencode and refseq are not included in chess

chess_se=ce[ce["tID"].isin(good_tIDs)].reset_index(drop=True)
gencode_merge_chess=chess_se.merge(right=gencode[gencode["type"]=="exon"].reset_index(drop=True),how="outer",on=["start","end"],indicator=True)

print("Out of a total of %d single-exon transcripts:\n\t%d are in CHESS\n\t%d are in GENCODE/RefSeq\n\t%d are shared between two annotations\n\t%d are found exclusively in CHESS\n\t%d are found exclusively in GENCODE/RefSeq"%(len(gencode_merge_chess),len(chess_se),len(gencode[gencode["type"]=="exon"]),len(gencode_merge_chess[gencode_merge_chess["_merge"]=="both"]),len(gencode_merge_chess[gencode_merge_chess["_merge"]=="left_only"]),len(gencode_merge_chess[gencode_merge_chess["_merge"]=="right_only"])))

# now let's have a look at the attirbutes of those genes/transcripts that are not included in CHESS
notInChess=gencode_merge_chess[gencode_merge_chess["_merge"]=="right_only"]
notInChess_genIDs=notInChess["attributes_y"].str.split("Parent=",expand=True)[1].str.strip("\n")
print(len(notInChess_genIDs))

notInChessTrans=gencode[~(gencode["type"].isin(["exon"]))].reset_index(drop=True)
notInChessTrans["ID"]=notInChessTrans["attributes"].str.split("ID=",expand=True)[1].str.split(";",expand=True)[0]
notInChessTrans=notInChessTrans[notInChessTrans["ID"].isin(notInChess_genIDs)].reset_index(drop=True)
notInChessTrans
for i in notInChess_genIDs:
    print(notInChessTrans[notInChessTrans["ID"]==i]["attributes"].iloc[0])
```

```python
# now onto actually producing the new gff

# first need to get the list of transcript and gene IDs to remove
outFP=open("./chess2.01.gff","w+")

count=0
with open("chess2.0.gff","r") as inFP:
    for line in inFP.readlines():
        if not line[0]=="#": # skip comments, otherwise proceed safely
            lineCols=line.split("\t")
            if lineCols[0]=="chrM":
                if lineCols[2]=="exon" or lineCols[2]=="CDS": # extract parent
                    parentID=lineCols[8].split("=")[-1].strip()
                    if not parentID in bad_tIDs:
                        outFP.write(line)
                elif lineCols[2]=="gene": # extract and compare geneID
                    geneID=lineCols[8].split("=")[1].split(";")[0]
                    if geneID in bad_gIDs:
                        count+=1
                    if not geneID in bad_gIDs or geneID in good_gIDs:
                        outFP.write(line)
                else: # extract transcript ID and compare
                    transID=lineCols[8].split("=")[1].split(";")[0]
                    if not transID in bad_tIDs:
                        outFP.write(line)
            else:
                outFP.write(line)
        else:
            if line[:7]=="##CHESS":
                outFP.write("##CHESS"+version+"\n")
            else:
                outFP.write(line)
            continue
print(count)
inFP.close()
outFP.close()

# then repeat for the other files:
# 1. chess_and_refseq
# 2. chess_assembly
# 3. recreate proteins
# 4. recreate genes
# 5. recreate transcripts
# 6. mapfile.txt
```

```python
# now let's deal with the proteins file

outFP=open("./chess2.01.protein.fa","w+")

count=0
allGood=False
with open("./chess2.0.protein.fa","r") as inFP:
    for line in inFP.readlines():
        if line[0]==">":
            chsID=line[1:].split(" ")[0]
            count+=1
            if not chsID in bad_tIDs:
                outFP.write(line)
                allGood=True
            else:
                allGood=False
                print(chsID)
        elif allGood:
            outFP.write(line)
        else:
            continue
        
inFP.close()
outFP.close()
```

```python
# now moving onto the mapfile.txt as the last stop

outFP=open("./mapfile2.01.txt","w+")

count=0
with open("./mapfile.txt","r") as inFP:
    for line in inFP.readlines():
        lineCols=line.split("\t")
        chsID=lineCols[1].strip()
        if chsID in bad_tIDs or not (not chsID in bad_gIDs or chsID in good_gIDs):
            continue
        outFP.write(line)
    
outFP.close()
inFP.close()
```

```python
# and now to fix the transcripts and genes files
outFP=open("./chess2.01.genes","w+")

count=0
removed=[]
countRemovedLines=0
with open("./chess2.0.genes","r") as inFP:
    for line in inFP.readlines():
        lineCols=line.split("\t")
        ids=lineCols[2].split(",")
        newIDS=[]
        for gid in ids:
            if gid in bad_gIDs and not gid in good_gIDs:
                count+=1
                removed.append(gid)
            else:
                newIDS.append(gid)
        if len(newIDS)==0: # remove line
            countRemovedLines+=1
            continue
        else:
            lineCols[2]=",".join(newIDS)
            outFP.write("\t".join(lineCols))

print(count)
print(countRemovedLines)
print(removed)
outFP.close()
inFP.close()
```

```python
print(len(set(bad_tIDs)))

outFP=open("./chess2.01.transcripts","w+")
count=0
removed=[]
countRemovedLines=0
with open("./chess2.0.transcripts","r") as inFP:
    for line in inFP.readlines():
        lineCols=line.split("\t")
        ids=lineCols[0].split(",")
        newIDS=[]
        for tid in ids:
            if tid in bad_tIDs:
                count+=1
                removed.append(tid)
            else:
                newIDS.append(tid)
        if len(newIDS)==0:
            countRemovedLines+=1
        else:
            lineCols[0]=",".join(newIDS)
            outFP.write("\t".join(lineCols))

print(count)
print(countRemovedLines)
print(removed)
outFP.close()
inFP.close()
```
