```python
import os
import re
from shutil import copyfile

version="2.03"
```

```python
outFP=open("./chess2.03.gff","w+")
vk=re.compile("(?:^[a-zA-Z_]*=|;[a-zA-Z_]*=)")
c=0
with open("./chess2.02.gff","r") as inFP:
    for line in inFP.readlines():
        if line[0]=="#":
            if line[:7]=="##CHESS":
                outFP.write("##CHESS"+version+"\n")
            else:
                outFP.write(line)
            continue
        else:
            att=line.split("\t")[-1]
            vals=re.split(vk,att)[1:]
            keys=re.findall(vk,att)
            old_vals=re.split(vk,att)[1:]
            errors=False
            for i in range(len(vals)):
                if keys[i].lower() in [";dbxref=","dbxref="]:
                    continue
                if "=" in vals[i]:
                    vals[i]=vals[i].replace("=","%3D")
                    errors=True
                if ";" in vals[i]:
                    vals[i]=vals[i].replace(";","%3B")
                    errors=True
                if "&" in vals[i]:
                    vals[i]=vals[i].replace("&","%26")
                    errors=True
            if errors:
                print(att)
                att="".join([x for y in zip(keys,vals) for x in y])
                print(att)
                print("\n======================\n")
                outFP.write("\t".join(line.split("\t")[:-1]+[att]))
            else:
                outFP.write(line)
                
outFP.close()
```
