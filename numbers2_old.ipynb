```python
import pandas as pd
import sys
import os

annFP="./chess2.0.gff"
genesFP="./chess2.0.genes"
transcriptsFP="./chess2.0.transcripts"
```

```python
listGenes=[]
def addGene(row):
    global listGenes
    ids=row["GFF_ID"].split(",")
    listGenes=listGenes+ids

gffCols=["seqid","source","type","start","end","score","strand","phase","attributes"]
df1=pd.read_csv(annFP,sep="\t",comment="#",names=gffCols)
df1t=df1[df1["type"]=="transcript"].reset_index(drop=True)
df1t["tID"]=df1t["attributes"].str.split("ID=",expand=True)[1].str.split(";",expand=True)[0]
df1t["i2"]=df1t["tID"].str.extract('(CHS.(\d)*.)',expand=True)[0]
df1t["gID"]=df1t["i2"].str.extract('(CHS.(\d)*)',expand=True)[0]
df1t.drop(["i2"],axis=1,inplace=True)

df1_genes=pd.read_csv(genesFP,sep="\t")
print("total number of genes: "+str(len(df1_genes)))
listGenes=[]
df1_genes.apply(lambda row: addGene(row),axis=1)
setTotalGenes=set(listGenes)
print("total number of genes: "+str(len(setTotalGenes)))
print("total number of transcripts: "+str(len(set(df1t["tID"]))))

print("number of protein_coding genes: "+str(len(df1_genes[df1_genes["Gene_Type"]=="protein_coding"])))
listGenes=[]
df1_genes[df1_genes["Gene_Type"]=="protein_coding"].apply(lambda row: addGene(row),axis=1)
setProteinCodingGenes=set(listGenes)
df1t_prot=df1t[df1t["gID"].isin(setProteinCodingGenes)].reset_index(drop=True)
print("number of protein_coding transcripts: "+str(len(df1t_prot)))

print("number of lncRNA genes: "+str(len(df1_genes[df1_genes["Gene_Type"]=="lncRNA"])))
listGenes=[]
df1_genes[df1_genes["Gene_Type"]=="lncRNA"].apply(lambda row: addGene(row),axis=1)
setlncRNAGenes=set(listGenes)
df1t_lncRNA=df1t[df1t["gID"].isin(setlncRNAGenes)].reset_index(drop=True)
print("number of lncRNA transcripts: "+str(len(df1t_lncRNA)))

print("number of antisense_RNA genes: "+str(len(df1_genes[df1_genes["Gene_Type"]=="antisense_RNA"])))
listGenes=[]
df1_genes[df1_genes["Gene_Type"]=="antisense_RNA"].apply(lambda row: addGene(row),axis=1)
setantisenseRNAGenes=set(listGenes)
df1t_antisenseRNA=df1t[df1t["gID"].isin(setantisenseRNAGenes)].reset_index(drop=True)
print("number of antisense_RNA transcripts: "+str(len(df1t_antisenseRNA)))

print("number of misc_RNA genes: "+str(len(df1_genes[df1_genes["Gene_Type"]=="misc_RNA"])))
listGenes=[]
df1_genes[df1_genes["Gene_Type"]=="misc_RNA"].apply(lambda row: addGene(row),axis=1)
setmiscRNAGenes=set(listGenes)
df1t_miscRNA=df1t[df1t["gID"].isin(setmiscRNAGenes)].reset_index(drop=True)
print("number of misc_RNA transcripts: "+str(len(df1t_miscRNA)))

# print("number of other genes: "+str(len(df1_genes)-(len(df1_genes[df1_genes["Gene_Type"]=="protein_coding"])+len(df1_genes[df1_genes["Gene_Type"]=="lncRNA"]))))
# print("number of other transcripts: "+str(len(set(df1t["tID"]))-(len(df1t_prot)+len(df1t_lncRNA))))

novel_genes=df1_genes[(df1_genes["Database"]=="CHESS")|(df1_genes["Database"]=="FANTOM")]
print("total number of novel genes: "+str(len(novel_genes)))
print("number of novel protein_coding genes: "+str(len(novel_genes[novel_genes["Gene_Type"]=="protein_coding"])))
listGenes=[]
novel_genes[novel_genes["Gene_Type"]=="protein_coding"].apply(lambda row: addGene(row),axis=1)
setNovelprotGenes=set(listGenes)
novel_transc_prot=df1t[df1t["gID"].isin(setNovelprotGenes)].reset_index(drop=True)
print("number of novel protein_coding transcripts: "+str(len(novel_transc_prot)))

print("number of novel lncRNA genes: "+str(len(novel_genes[novel_genes["Gene_Type"]=="lncRNA"])))
listGenes=[]
novel_genes[novel_genes["Gene_Type"]=="lncRNA"].apply(lambda row: addGene(row),axis=1)
setNovellncRNAGenes=set(listGenes)
novel_transc_lncRNA=df1t[df1t["gID"].isin(setNovellncRNAGenes)].reset_index(drop=True)
print("number of novel lncRNA transcripts: "+str(len(novel_transc_lncRNA)))

print("number of novel antisense genes: "+str(len(novel_genes[novel_genes["Gene_Type"]=="antisense_RNA"])))
listGenes=[]
novel_genes[novel_genes["Gene_Type"]=="antisense_RNA"].apply(lambda row: addGene(row),axis=1)
setNovelantiRNAGenes=set(listGenes)
novel_transc_antiRNA=df1t[df1t["gID"].isin(setNovelantiRNAGenes)].reset_index(drop=True)
print("number of novel antisense transcripts: "+str(len(novel_transc_antiRNA)))

print("number of novel misc genes: "+str(len(novel_genes[novel_genes["Gene_Type"]=="misc_RNA"])))
listGenes=[]
novel_genes[novel_genes["Gene_Type"]=="misc_RNA"].apply(lambda row: addGene(row),axis=1)
setNovelmiscRNAGenes=set(listGenes)
novel_transc_miscRNA=df1t[df1t["gID"].isin(setNovelmiscRNAGenes)].reset_index(drop=True)
print("number of novel misc transcripts: "+str(len(novel_transc_miscRNA)))
```

```python
# need to find polycistronic genes/transcripts
# needs to be done based on refseq
dontCount=["CHS.59388.1",
            "CHS.51943.4",
            "CHS.10180.8",
            "CHS.26116.2",
            "CHS.48289.2",
            "CHS.48289.3",
            "CHS.48749.2",
            "CHS.48928.2"]

listGenes=[]
def addGene(row):
    global listGenes
    ids=row["GFF_ID"].split(",")
    listGenes=listGenes+ids

gffCols=["seqid","source","type","start","end","score","strand","phase","attributes"]
df1=pd.read_csv(annFP,sep="\t",comment="#",names=gffCols)
df1t=df1[df1["type"]=="transcript"].reset_index(drop=True)
df1t["tID"]=df1t["attributes"].str.split("ID=",expand=True)[1].str.split(";",expand=True)[0]
df1t["i2"]=df1t["tID"].str.extract('(CHS.(\d)*.)',expand=True)[0]
df1t["gID"]=df1t["i2"].str.extract('(CHS.(\d)*)',expand=True)[0]
df1t.drop(["i2"],axis=1,inplace=True)

df1_genes=pd.read_csv(genesFP,sep="\t")
print("total number of genes: "+str(len(df1_genes)))
listGenes=[]
df1_genes.apply(lambda row: addGene(row),axis=1)
setTotalGenes=set(listGenes)
print("total number of genes: "+str(len(setTotalGenes)))
print("total number of transcripts: "+str(len(set(df1t[~df1t["tID"].isin(dontCount)]["tID"]))))

print("number of protein_coding genes: "+str(len(df1_genes[df1_genes["Gene_Type"]=="protein_coding"])))
listGenes=[]
df1_genes[df1_genes["Gene_Type"]=="protein_coding"].apply(lambda row: addGene(row),axis=1)
setProteinCodingGenes=set(listGenes)
print("number of protein_coding genes: "+str(len(setProteinCodingGenes)))
# df1t_prot=df1t[(df1t["gID"].isin(setProteinCodingGenes))&~(df1t["tID"].isin(dontCount))].reset_index(drop=True)
# print("number of protein_coding transcripts: "+str(len(df1t_prot)))

print("number of lncRNA genes: "+str(len(df1_genes[df1_genes["Gene_Type"]=="lncRNA"])))
listGenes=[]
df1_genes[df1_genes["Gene_Type"]=="lncRNA"].apply(lambda row: addGene(row),axis=1)
setlncRNAGenes=set(listGenes)
print("number of lncRNA genes: "+str(len(setlncRNAGenes)))

print("number of antisense_RNA genes: "+str(len(df1_genes[df1_genes["Gene_Type"]=="antisense_RNA"])))
listGenes=[]
df1_genes[df1_genes["Gene_Type"]=="antisense_RNA"].apply(lambda row: addGene(row),axis=1)
setantisenseRNAGenes=set(listGenes)
print("number of antisense_RNA genes: "+str(len(setantisenseRNAGenes)))

print("number of misc_RNA genes: "+str(len(df1_genes[df1_genes["Gene_Type"]=="misc_RNA"])))
listGenes=[]
df1_genes[df1_genes["Gene_Type"]=="misc_RNA"].apply(lambda row: addGene(row),axis=1)
setmiscRNAGenes=set(listGenes)
print("number of misc_RNA genes: "+str(len(setmiscRNAGenes)))
```
