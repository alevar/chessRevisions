```python
import os
import re
import sys
import csv
import pandas as pd
import numpy as np

gff3Cols=["seqid","source","type","start","end","score","strand","phase","attributes"]
```

```python
# this notebook is meant to investigate where exactly the missing transcripts are coming from
# missing transcripts were identified by Alaina as part of the intron chain analysis
```

```python
# first load the missing transcript IDS

missing=[]
with open("./missing.novel.chess2.1","r") as inFP:
    for line in inFP.readlines():
        missing.append(line.strip())
inFP.close()

print(len(missing))
print(len(set(missing)))
```

```python
df2_1=pd.read_csv("./chess2.1.gff",sep="\t",names=gff3Cols)
df2_1=df2_1[df2_1["type"]=="transcript"].reset_index(drop=True)
df2_1.dropna(inplace=True,axis=0)
df2_1.reset_index(inplace=True,drop=True)
df2_1["start"]=df2_1["start"].astype(int)
df2_1["end"]=df2_1["end"].astype(int)
df2_1["id"]=df2_1.attributes.str.split("ID=",expand=True)[1].str.split(";",expand=True)[0]
print(len(df2_1[df2_1["id"].isin(missing)]))
df2_1[df2_1["id"].isin(missing)]
```
